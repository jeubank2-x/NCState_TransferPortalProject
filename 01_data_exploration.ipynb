```python
import pandas as pd

ratings = pd.read_csv("data/raw/player_ratings.csv")
portal = pd.read_csv("data/raw/portal_players.csv")

print(f"Player ratings rows: {len(ratings)}")
print(f"Portal players rows: {len(portal)}")
```

```python
print("=== PLAYER RATINGS COLUMNS ===")
print(list(ratings.columns))

print("\n=== FIRST 3 ROWS OF RATINGS ===")
print(ratings.head(3))

print("\n=== PORTAL PLAYERS COLUMNS ===")
print(list(portal.columns))

print("\n=== FIRST 3 ROWS OF PORTAL ===")
print(portal.head(3))
```

```python

print("=== POSITIONS IN PORTAL ===")
print(portal['Position'].value_counts())

print("\n=== YEARS OF ELIGIBILITY ===")
print(portal['Year'].value_counts())


print("\n=== ANY MISSING DATA? ===")
print(portal.isnull().sum())

```
